```python
# ============================================================
# ANNITIA CHALLENGE — IMPROVED SOLUTION
# Uses proper survival analysis (C-index optimization)
# with longitudinal feature engineering on NIT trajectories
# ============================================================

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold

# Install scikit-survival if needed
# pip install scikit-survival

from sksurv.util import Surv
from sksurv.linear_model import CoxPHSurvivalAnalysis, CoxnetSurvivalAnalysis
from sksurv.ensemble import RandomSurvivalForest
from sksurv.metrics import concordance_index_censored

np.random.seed(42)

# ============================================================
# 1. LOAD DATA
# ============================================================
print("Loading data...")
train = pd.read_csv("/kaggle/input/datasets/qamarmath/annitia-data-challenge/trainANNITIA.csv")
test  = pd.read_csv("/kaggle/input/datasets/qamarmath/annitia-data-challenge/test ANNITIA.csv")
print(f"Train: {train.shape}, Test: {test.shape}")

# ============================================================
# 2. LONGITUDINAL FEATURE ENGINEERING
#    The whole point of ANNITIA is NIT trajectories over time.
#    Extract slope, last value, max, min, std, n_obs for each
#    repeated measurement series.
# ============================================================
print("\nEngineering longitudinal features...")

# All repeated measurement prefixes (v1..v22)
NIT_PREFIXES = [
    'bmi',
    'ALT', 'AST', 'GGT', 'bilirubin', 'platelets',
    'glycemia', 'triglycerides', 'total_cholesterol',
    'FibroTest',
    'fibs_stiffness_med_BM_1',   # FibroScan
    'fibs_CAP_BM_1',             # FibroScan CAP (steatosis)
    'SSI',                        # Aixplorer shear wave
]

def extract_longitudinal_features(df):
    """
    For each measurement series, compute:
      - last observed value (most recent visit)
      - first value (baseline)
      - max, min, mean, std
      - slope (linear regression coefficient over time using Age columns)
      - number of observations
    """
    age_cols = sorted([c for c in df.columns if c.startswith('Age_v')],
                      key=lambda x: int(x.split('_v')[1]))
    
    features = {}
    
    for prefix in NIT_PREFIXES:
        # Find all visit columns for this prefix
        visit_cols = sorted(
            [c for c in df.columns if c.startswith(prefix + '_v') or
             (prefix == 'bmi' and c.startswith('BMI_v'))],
            key=lambda x: int(x.split('_v')[1])
        )
        # Also handle lowercase bmi
        if prefix == 'bmi':
            visit_cols = sorted(
                [c for c in df.columns if c.lower().startswith('bmi_v')],
                key=lambda x: int(x.split('_v')[1])
            )
        
        if not visit_cols:
            continue
        
        vals = df[visit_cols].values.astype(float)     # (n_patients, n_visits)
        ages = df[age_cols[:len(visit_cols)]].values.astype(float)
        
        n = vals.shape[0]
        last_val   = np.full(n, np.nan)
        first_val  = np.full(n, np.nan)
        max_val    = np.full(n, np.nan)
        min_val    = np.full(n, np.nan)
        mean_val   = np.full(n, np.nan)
        std_val    = np.full(n, np.nan)
        slope      = np.full(n, np.nan)
        n_obs      = np.zeros(n)
        
        for i in range(n):
            v = vals[i]
            a = ages[i]
            mask = ~np.isnan(v) & ~np.isnan(a)
            obs_v = v[mask]
            obs_a = a[mask]
            
            if len(obs_v) == 0:
                continue
            
            n_obs[i]   = len(obs_v)
            last_val[i]  = obs_v[-1]
            first_val[i] = obs_v[0]
            max_val[i]   = obs_v.max()
            min_val[i]   = obs_v.min()
            mean_val[i]  = obs_v.mean()
            std_val[i]   = obs_v.std() if len(obs_v) > 1 else 0.0
            
            # Slope via least squares (only meaningful with 2+ obs)
            if len(obs_v) >= 2:
                A = np.vstack([obs_a, np.ones(len(obs_a))]).T
                result = np.linalg.lstsq(A, obs_v, rcond=None)
                slope[i] = result[0][0]
        
        p = prefix.replace('fibs_stiffness_med_BM_1', 'fibroscan')\
                   .replace('fibs_CAP_BM_1', 'cap')\
                   .replace('total_cholesterol', 'chol')\
                   .lower()
        
        features[f'{p}_last']   = last_val
        features[f'{p}_first']  = first_val
        features[f'{p}_max']    = max_val
        features[f'{p}_min']    = min_val
        features[f'{p}_mean']   = mean_val
        features[f'{p}_std']    = std_val
        features[f'{p}_slope']  = slope
        features[f'{p}_nobs']   = n_obs
        features[f'{p}_change'] = last_val - first_val   # absolute change
    
    return pd.DataFrame(features, index=df.index)


train_long = extract_longitudinal_features(train)
test_long  = extract_longitudinal_features(test)
print(f"Longitudinal features: {train_long.shape[1]}")

# ============================================================
# 3. STATIC / BASELINE FEATURES
# ============================================================
STATIC_COLS = ['gender', 'T2DM', 'Hypertension', 'Dyslipidaemia',
               'bariatric_surgery', 'Age_v1']

# Add bariatric surgery age if present
for col in ['bariatric_surgery_age']:
    if col in train.columns:
        STATIC_COLS.append(col)

train_static = train[STATIC_COLS].copy()
test_static  = test[STATIC_COLS].copy()

# Follow-up duration
age_cols_all = sorted([c for c in train.columns if c.startswith('Age_v')],
                      key=lambda x: int(x.split('_v')[1]))
train_static['follow_up_yrs'] = (
    train[age_cols_all].max(axis=1) - train['Age_v1']
)
age_cols_test = sorted([c for c in test.columns if c.startswith('Age_v')],
                       key=lambda x: int(x.split('_v')[1]))
test_static['follow_up_yrs'] = (
    test[age_cols_test].max(axis=1) - test['Age_v1']
)

# ============================================================
# 4. COMBINE ALL FEATURES
# ============================================================
X_train = pd.concat([train_static.reset_index(drop=True),
                     train_long.reset_index(drop=True)], axis=1)
X_test  = pd.concat([test_static.reset_index(drop=True),
                     test_long.reset_index(drop=True)], axis=1)

# Align columns
X_test = X_test.reindex(columns=X_train.columns, fill_value=np.nan)
print(f"Total features: {X_train.shape[1]}")

# ============================================================
# 5. BUILD SURVIVAL TARGETS
#    sksurv expects a structured array: (event: bool, time: float)
#    Time = age_at_event - age_at_baseline  (years)
#    For censored patients, time = last_observed_age - baseline_age
# ============================================================
print("\nBuilding survival targets...")

def build_survival_target(df, event_col, age_occur_col, age_baseline_col='Age_v1'):
    """
    Returns sksurv structured array.
    Excludes rows with unknown event status (NaN in event_col).
    Returns also a boolean mask for which rows are included.
    """
    age_cols = sorted([c for c in df.columns if c.startswith('Age_v')],
                      key=lambda x: int(x.split('_v')[1]))
    last_age = df[age_cols].max(axis=1)
    baseline = df[age_baseline_col]
    
    valid_mask = df[event_col].notna()
    df_v = df[valid_mask].copy()
    last_v = last_age[valid_mask]
    base_v = baseline[valid_mask]
    
    events = df_v[event_col].astype(bool).values
    times  = np.where(
        events & df_v[age_occur_col].notna(),
        df_v[age_occur_col] - base_v,
        last_v - base_v
    )
    
    # Clip negative / zero times (shouldn't happen but just in case)
    times = np.maximum(times, 0.01)
    
    y = Surv.from_arrays(event=events, time=times)
    return y, valid_mask.values

y_hep,   mask_hep   = build_survival_target(train,
    'evenements_hepatiques_majeurs', 'evenements_hepatiques_age_occur')
y_death, mask_death = build_survival_target(train,
    'death', 'death_age_occur')

X_train_hep   = X_train[mask_hep].reset_index(drop=True)
X_train_death = X_train[mask_death].reset_index(drop=True)

print(f"Hepatic model — patients: {len(y_hep)}, events: {y_hep['event'].sum()}")
print(f"Death model   — patients: {len(y_death)}, events: {y_death['event'].sum()}")

# ============================================================
# 6. PREPROCESSING PIPELINE
# ============================================================
def make_preprocessor():
    return Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler()),
    ])

# ============================================================
# 7. SURVIVAL MODEL TRAINING
#    Strategy:
#    - CoxnetSurvivalAnalysis (elastic-net penalized Cox) — handles
#      high-dimensional, collinear features; good for hepatic model
#    - RandomSurvivalForest — handles non-linearities, works well
#      even with very few events
#    - Ensemble: average risk scores from both
# ============================================================
print("\nTraining survival models...")

def train_survival_ensemble(X, y, label="model"):
    """
    Train CoxNet + RSF, cross-validate C-index, return trained models
    and preprocessor fitted on full data.
    """
    print(f"\n--- {label} ---")
    print(f"  Patients: {len(y)}, Events: {y['event'].sum()}")
    
    prep = make_preprocessor()
    X_proc = prep.fit_transform(X)
    
    # ---- CoxNet (Elastic Net Cox) ----
    # alpha_min_ratio controls regularization strength range
    cox = CoxnetSurvivalAnalysis(
        l1_ratio=0.5,           # mix of L1 and L2
        fit_baseline_model=True,
        alpha_min_ratio=0.1,
        max_iter=1000,
        normalize=False,        # already scaled
    )
    cox.fit(X_proc, y)
    
    # ---- Random Survival Forest ----
    # n_estimators higher = more stable, especially with few events
    rsf = RandomSurvivalForest(
        n_estimators=3000,
        min_samples_split=10,
        min_samples_leaf=5,
        max_features='sqrt',
        n_jobs=-1,
        random_state=42,
    )
    rsf.fit(X_proc, y)
    
    # ---- Cross-validated C-index ----
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    cox_cis, rsf_cis = [], []
    
    for fold, (tr_idx, va_idx) in enumerate(kf.split(X_proc)):
        X_tr, X_va = X_proc[tr_idx], X_proc[va_idx]
        y_tr, y_va = y[tr_idx], y[va_idx]
        
        # Skip fold if no events in val
        if y_va['event'].sum() == 0:
            continue
        
        try:
            cox_f = CoxnetSurvivalAnalysis(
                l1_ratio=0.5, fit_baseline_model=True,
                alpha_min_ratio=0.1, max_iter=1000, normalize=False
            )
            cox_f.fit(X_tr, y_tr)
            risk_cox = cox_f.predict(X_va)
            ci_cox, _, _, _, _ = concordance_index_censored(
                y_va['event'], y_va['time'], risk_cox)
            cox_cis.append(ci_cox)
        except Exception:
            pass
        
        try:
            rsf_f = RandomSurvivalForest(
                n_estimators=200, min_samples_split=10,
                min_samples_leaf=5, max_features='sqrt',
                n_jobs=-1, random_state=42
            )
            rsf_f.fit(X_tr, y_tr)
            risk_rsf = rsf_f.predict(X_va)
            ci_rsf, _, _, _, _ = concordance_index_censored(
                y_va['event'], y_va['time'], risk_rsf)
            rsf_cis.append(ci_rsf)
        except Exception:
            pass
    
    if cox_cis:
        print(f"  CoxNet CV C-index: {np.mean(cox_cis):.4f} ± {np.std(cox_cis):.4f}")
    if rsf_cis:
        print(f"  RSF    CV C-index: {np.mean(rsf_cis):.4f} ± {np.std(rsf_cis):.4f}")
    
    return prep, cox, rsf


prep_hep,   cox_hep,   rsf_hep   = train_survival_ensemble(
    X_train_hep, y_hep, label="Hepatic events")
prep_death, cox_death, rsf_death = train_survival_ensemble(
    X_train_death, y_death, label="Death")


# ============================================================
# 8. PREDICT RISK SCORES ON TEST SET
#    C-index is based on ranking: higher score = higher risk.
#    We ensemble CoxNet + RSF by rank-averaging.
# ============================================================
print("\nGenerating predictions...")

def rank_normalize(arr):
    """Convert raw scores to rank-normalized [0,1] scores."""
    from scipy.stats import rankdata
    return rankdata(arr) / len(arr)

def predict_risk(X_test, prep, cox_model, rsf_model):
    X_proc = prep.transform(X_test)
    risk_cox = cox_model.predict(X_proc)
    risk_rsf = rsf_model.predict(X_proc)
    # Rank-normalize then average (avoids scale differences)
    risk_ensemble = (rank_normalize(risk_cox) + rank_normalize(risk_rsf)) / 2
    return risk_ensemble

risk_hepatic = predict_risk(X_test, prep_hep,   cox_hep,   rsf_hep)
risk_death   = predict_risk(X_test, prep_death, cox_death, rsf_death)

print(f"Hepatic risk — min: {risk_hepatic.min():.4f}, max: {risk_hepatic.max():.4f}")
print(f"Death risk   — min: {risk_death.min():.4f}, max: {risk_death.max():.4f}")

# ============================================================
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import random
import copy
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sksurv.metrics import concordance_index_censored

# 1. Reproducibility
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    print(f"✅ Seed set to: {seed}")

set_seed(42)

# 2. DeepSurv Architecture (Regularized)
class DeepSurv(nn.Module):
    def __init__(self, input_dim):
        super(DeepSurv, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_dim, 32),
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.Dropout(0.4), # Helps prevent memorization
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 1)
        )
        
    def forward(self, x):
        return torch.clamp(self.network(x), min=-10, max=10)

# 3. Loss Function
def cox_loss(risk_pred, event, time):
    idx = torch.argsort(time, descending=True)
    risk_pred = risk_pred[idx]
    event = event[idx]
    max_risk = torch.max(risk_pred)
    exp_risk = torch.exp(risk_pred - max_risk)
    log_sum_exp_risk = torch.log(torch.cumsum(exp_risk, dim=0) + 1e-8) + max_risk
    return -torch.sum(event * (risk_pred - log_sum_exp_risk)) / (torch.sum(event) + 1e-8)

# 4. Training with Detailed Diagnostics
def train_and_diagnose(X, y, label="Model"):
    print(f"\n" + "="*80)
    print(f"🔍 DIAGNOSTIC TRAINING: {label}")
    print("="*80)
    
    # Preprocessing
    imputer = SimpleImputer(strategy='median')
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(imputer.fit_transform(X)).astype(np.float32)
    
    X_tr, X_va, y_tr, y_va = train_test_split(
        X_scaled, y, test_size=0.2, random_state=42, stratify=y['event']
    )
    
    x_tr_t = torch.from_numpy(X_tr)
    t_tr_t = torch.from_numpy(y_tr['time'].astype(np.float32).copy())
    e_tr_t = torch.from_numpy(y_tr['event'].astype(np.float32).copy())
    
    x_va_t = torch.from_numpy(X_va)
    t_va_t = torch.from_numpy(y_va['time'].astype(np.float32).copy())
    e_va_t = torch.from_numpy(y_va['event'].astype(np.float32).copy())

    model = DeepSurv(x_tr_t.shape[1])
    # Lower Learning Rate to prevent aggressive memorization
    optimizer = torch.optim.Adam(model.parameters(), lr=0.0005, weight_decay=0.1)
    
    best_ci_va = 0
    best_model_wts = copy.deepcopy(model.state_dict())
    
    print(f"{'Epoch':<8} | {'Tr-Loss':<10} | {'Va-Loss':<10} | {'Tr-C-Idx':<10} | {'Va-C-Idx':<10} | {'Risk-Std':<10}")
    print("-" * 80)

    for epoch in range(401):
        model.train()
        optimizer.zero_grad()
        risk_tr = model(x_tr_t).squeeze()
        loss_tr = cox_loss(risk_tr, e_tr_t, t_tr_t)
        loss_tr.backward()
        optimizer.step()
        
        if epoch % 50 == 0:
            model.eval()
            with torch.no_grad():
                risk_va = model(x_va_t).squeeze()
                loss_va = cox_loss(risk_va, e_va_t, t_va_t)
                
                # C-index for both
                ci_tr = concordance_index_censored(y_tr['event'], y_tr['time'], risk_tr.detach().numpy())[0]
                ci_va = concordance_index_censored(y_va['event'], y_va['time'], risk_va.numpy())[0]
                
                # Risk Score Diversity (Lower std means model is becoming too confident/biased)
                risk_std = torch.std(risk_va).item()
                
                if ci_va > best_ci_va:
                    best_ci_va = ci_va
                    best_model_wts = copy.deepcopy(model.state_dict())

                print(f"{epoch:<8} | {loss_tr.item():<10.4f} | {loss_va.item():<10.4f} | {ci_tr:<10.4f} | {ci_va:<10.4f} | {risk_std:<10.4f}")

    # Summary Analysis
    model.load_state_dict(best_model_wts)
    print("-" * 80)
    print(f"✅ Final Best Validation C-Index: {best_ci_va:.4f}")
    
    # Final Memorization Check
    # If gap is > 0.15, you are memorizing!
    gap = abs(ci_tr - best_ci_va)
    if gap > 0.15:
        print(f"⚠️  WARNING: High Memorization detected! (Gap: {gap:.4f})")
    else:
        print(f"👍 Model is learning patterns well. (Gap: {gap:.4f})")
        
    return model, imputer, scaler

# 5. Run Models
model_hep, imp_hep, scl_hep = train_and_diagnose(X_train_hep, y_hep, "HEPATIC EVENT")
model_death, imp_death, scl_death = train_and_diagnose(X_train_death, y_death, "DEATH")

# 6. Prediction
def get_predictions(model, X_test, imputer, scaler):
    model.eval()
    X_proc = scaler.transform(imputer.transform(X_test)).astype(np.float32)
    with torch.no_grad():
        return model(torch.from_numpy(X_proc)).squeeze().numpy()

risk_hep = get_predictions(model_hep, X_test, imp_hep, scl_hep)
risk_death = get_predictions(model_death, X_test, imp_death, scl_death)

# Submission
submission = pd.DataFrame({
    'trustii_id': test['trustii_id'].values,
    'risk_hepatic_event': risk_hep,
    'risk_death': risk_death
})
submission.to_csv('submission_diagnostic.csv', index=False)
print("\n📁 Submission saved as 'submission_diagnostic.csv'")
```
